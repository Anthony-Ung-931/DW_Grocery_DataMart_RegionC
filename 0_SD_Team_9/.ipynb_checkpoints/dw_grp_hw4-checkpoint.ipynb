```python
import csv
from decimal import Decimal
import random
import datetime
from datetime import date
import sqlite3 as lite
import math
from decimal import Decimal

# Database setup
con = sqlite3.connect('grocery_store.db')
cur = con.cursor()
cur.execute('DROP TABLE IF EXISTS transactions')
cur.execute('CREATE TABLE IF NOT EXISTS transactions(date1 TEXT, customerID INT, sku INT, salePrice REAL, itemsLeft INT, co INT)')

csv.register_dialect('piper', delimiter='|', quoting=csv.QUOTE_NONE)

#team 9 Parameters
price_multiplier = 1.07
total_items_bought = 0
customer_count = 0
transactionsMade = 0

# Initial Inventory Supply
milkSupply = 1110  # 1.5-day supply based on daily sales
cerealSupply = 1152  # 3-day supply
babySupply = 630  # 3-day supply
diaperSupply = 528  # 3-day supply
breadSupply = 1575  # 3-day supply
peanutSupply = 315  # 3-day supply
jellySupply = 426  # 3-day supply
nonSpecialSupply = 75  # 3-day supply per non-special item

manufacturer_list = []
product_name_list = []
size_list = []
item_type_list = []
skus = []

milk_price = []
milk_sku = []
milkStock = []
milkContainer = []

cereal_price = []
cereal_sku = []
cerealStock = []
cerealContainer = []

baby_food_price = []
baby_food_sku = []
babyStock = []
babyContainer = []

diapers_price = []
diapers_sku = []
diaperStock = []
diaperContainer = []

bread_price = []
bread_sku = []
breadStock = []
breadContainer = []

peanut_butter_price = []
peanut_butter_sku = []
peanutStock = []
peanutContainer = []

jelly_jam_price = []
jelly_jam_sku = []
jellyStock = []
jellyContainer = []

all_items_price = []
all_items = []
allStock = []
allContainer = []

# Load products from file

with open('Products1.txt', 'r', encoding='ISO-8859-1') as csvfile:
    for row in csv.DictReader(csvfile, dialect='piper'):
        nosign = row['BasePrice']
        nosign = float(Decimal(nosign.strip('$')))
        nosign = round(nosign*price_multiplier, 2)
        manufacturer_list.append(row.get('Manufacturer'))
        product_name_list.append(row.get('Product Name'))
        size_list.append(row.get('Size'))
        item_type_list.append(row.get('itemType'))
        skus.append(row['SKU'])
        if (row['itemType'] == 'Milk'):
            milk_price.append(nosign)
            milk_sku.append(row['SKU'])
            milkStock.append(milkSupply)
            milkContainer.append(milkSupply / 12)
        elif (row['itemType'] == 'Cereal'):
            cereal_price.append(nosign)
            cereal_sku.append(row['SKU'])
            cerealStock.append(cerealSupply)
            cerealContainer.append(cerealSupply / 12)
        elif (row['itemType'] == 'Baby Food'):
            baby_food_price.append(nosign)
            baby_food_sku.append(row['SKU'])
            babyStock.append(babySupply)
            babyContainer.append(babySupply / 12)
        elif (row['itemType'] == 'Diapers'):
            diapers_price.append(nosign)
            diapers_sku.append(row['SKU'])
            diaperStock.append(diaperSupply)
            diaperContainer.append(diaperSupply / 12)
        elif (row['itemType'] == 'Bread'):
            bread_price.append(nosign)
            bread_sku.append(row['SKU'])
            breadStock.append(breadSupply)
            breadContainer.append(breadSupply / 12)
        elif (row['itemType'] == 'Peanut Butter'):
            peanut_butter_price.append(nosign)
            peanut_butter_sku.append(row['SKU'])
            peanutStock.append(peanutSupply)
            peanutContainer.append(peanutSupply / 12)
        elif (row['itemType'] == 'Jelly/Jam'):
            jelly_jam_price.append(nosign)
            jelly_jam_sku.append(row['SKU'])
            jellyStock.append(jellySupply)
            jellyContainer.append(jellySupply / 12)
        else:
            all_items_price.append(nosign)
            all_items.append(row.get('SKU'))
            allStock.append(nonSpecialSupply)
            allContainer.append(nonSpecialSupply / 12)

def orderProducts(date):
    for i in range(len(milkStock)):
        if(milkStock[i] < milkSupply):
            amountNeeded = milkSupply - milkStock[i]
            containersNeeded = math.ceil(amountNeeded / 12)
            milkStock[i] = milkStock[i] + containersNeeded * 12
            milkContainer[i] += containersNeeded
    if(date == 2 or date == 4 or date == 6):
        for i in range(len(cerealStock)):
            if(cerealStock[i] < cerealSupply):
                amountNeeded = cerealSupply - cerealStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                cerealStock[i] = cerealStock[i] + containersNeeded * 12
                cerealContainer[i] += containersNeeded
        for i in range(len(babyStock)):
            if(babyStock[i] < babySupply):
                amountNeeded = babySupply - babyStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                babyStock[i] = babyStock[i] + containersNeeded * 12
                babyContainer[i] += containersNeeded
        for i in range(len(diaperStock)):
            if(diaperStock[i] < diaperSupply):
                amountNeeded = diaperSupply - diaperStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                diaperStock[i] = diaperStock[i] + containersNeeded * 12
                diaperContainer[i] += containersNeeded
        for i in range(len(breadStock)):
            if(breadStock[i] < breadSupply):
                amountNeeded = breadSupply - breadStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                breadStock[i] = breadStock[i] + containersNeeded * 12
                breadContainer[i] += containersNeeded
        for i in range(len(peanutStock)):
            if(peanutStock[i] < peanutSupply):
                amountNeeded = peanutSupply - peanutStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                peanutStock[i] = peanutStock[i] + containersNeeded * 12
                peanutContainer[i] += containersNeeded
        for i in range(len(jellyStock)):
            if(jellyStock[i] < jellySupply):
                amountNeeded = jellySupply - jellyStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                jellyStock[i] = jellyStock[i] + containersNeeded * 12
                jellyContainer[i] += containersNeeded
        for i in range(len(allStock)):
            if(allStock[i] < nonSpecialSupply):
                amountNeeded = nonSpecialSupply - allStock[i]
                containersNeeded = math.ceil(amountNeeded / 12)
                allStock[i] = allStock[i] + containersNeeded * 12
                allContainer[i] += containersNeeded


def write_record(current_date, customer_number, SKU, price, items_Left, containers_Bought):
    global total_items_bought
    global transactionsMade
    total_items_bought = total_items_bought + 1
    transactionsMade += 1
    if ((transactionsMade % 20000) == 0):
        con.commit()
        print("Commited", transactionsMade)
    try:
        cur.execute("INSERT INTO transactions VALUES (?,?,?,?,?,?)",
            (str(current_date), customer_number, SKU, price, items_Left, containers_Bought))
    except lite.OperationalError as err:
        print("insert error: %s", err)

#Everything below this was courtesy of Leonchuck
def get_milk_sku_and_price():
    random_index = random.randrange(len(milk_sku))
    running = True
    while(running):
        if(milkStock[random_index] != 0):
            running = False
            milkStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return milk_sku[random_index], milk_price[random_index], milkStock[random_index], milkContainer[random_index]

def cereal_sku_and_price():
    random_index = random.randrange(len(cereal_sku))
    running = True
    while(running):
        if(cerealStock[random_index] != 0):
            running = False
            cerealStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return cereal_sku[random_index], cereal_price[random_index], cerealStock[random_index], cerealContainer[random_index]

def baby_food_sku_and_price():
    random_index = random.randrange(len(baby_food_sku))
    running = True
    while(running):
        if(babyStock[random_index] != 0):
            running = False
            babyStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return baby_food_sku[random_index], baby_food_price[random_index], babyStock[random_index], babyContainer[random_index]

def diapers_sku_and_price():
    random_index = random.randrange(len(diapers_sku))
    running = True
    while(running):
        if(diaperStock[random_index] != 0):
            running = False
            diaperStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return diapers_sku[random_index], diapers_price[random_index], diaperStock[random_index], diaperContainer[random_index]

def bread_sku_and_price():
    random_index = random.randrange(len(bread_sku))
    running = True
    while(running):
        if(breadStock[random_index] != 0):
            running = False
            breadStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return bread_sku[random_index], bread_price[random_index], breadStock[random_index], breadContainer[random_index]

def peanut_butter_sku_and_price():
    random_index = random.randrange(len(peanut_butter_sku))
    running = True
    while(running):
        if(peanutStock[random_index] != 0):
            running = False
            peanutStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return peanut_butter_sku[random_index], peanut_butter_price[random_index], peanutStock[random_index], peanutContainer[random_index]

def jelly_jam_sku_and_price():
    random_index = random.randrange(len(jelly_jam_sku))
    running = True
    while(running):
        if(jellyStock[random_index] != 0):
            running = False
            jellyStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))
    return jelly_jam_sku[random_index], jelly_jam_price[random_index], jellyStock[random_index], jellyContainer[random_index]

def get_random_item_sku_and_price():
    random_index = random.randrange(len(all_items))
    running = True
    while(running):
        if(allStock[random_index] != 0):
            running = False
            allStock[random_index] -= 1
        else:
            random_index = random.randrange(len(all_items))

    return all_items[random_index], all_items_price[random_index], allStock[random_index], allContainer[random_index]

customers_low = 1050
customers_high = 1080
weekend_increase = 50
maximum_items = 70
simulation_start_date = date(2024, 1, 1)
simulation_end_date = date(2024, 12, 31)
current_date = simulation_start_date
daily_customers = 0
my_items = 0
total_customers = 0



for iday in range(0,366):
    increase = 0
    orderProducts(current_date.weekday())
    if current_date.weekday() >= 5:
        increase = weekend_increase

    daily_customers = random.randint(customers_low + increase, customers_high + increase)
    customer_number = 1
    while customer_number <= daily_customers:
        customer_count = customer_count + 1
        my_items = random.randint(1, maximum_items)
        k = 0
        if random.randint(1,100) <= 70:
            SKU, price, amountLeft, containersBought = get_milk_sku_and_price()
            write_record(current_date,total_customers,SKU,price, amountLeft, containersBought)
            k += 1

            if random.randint(1,100) <= 50:
                SKU, price, amountLeft, containersBought = cereal_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k += 1

        else:
            if random.randint(1,100) <= 5:
                SKU, price, amountLeft, containersBought = cereal_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k += 1

        if random.randint(1,100) <= 20:
            SKU, price, amountLeft, containersBought = baby_food_sku_and_price()
            write_record(current_date,total_customers,SKU,price, amountLeft, containersBought)
            k += 1

            if random.randint(1,100) <= 80:
                SKU, price, amountLeft, containersBought = diapers_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k += 1

        else:
            if random.randint(1,100) <= 1:
                SKU, price, amountLeft, containersBought = diapers_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k +=1

        if random.randint(1,100) <= 50:
            SKU, price, amountLeft, containersBought = bread_sku_and_price()
            write_record(current_date,total_customers,SKU,price, amountLeft, containersBought)
            k += 1

        if random.randint(1,100) <= 10:
            SKU, price, amountLeft, containersBought = peanut_butter_sku_and_price()
            write_record(current_date,total_customers,SKU,price, amountLeft, containersBought)
            k += 1

            if random.randint(1,100) <= 90:
                SKU, price, amountLeft, containersBought = jelly_jam_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k += 1

        else:
            if random.randint(1,100) <= 5:
                SKU, price, amountLeft, containersBought = jelly_jam_sku_and_price()
                write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
                k += 1

        while k < my_items:
            SKU, price, amountLeft, containersBought = get_random_item_sku_and_price()
            write_record(current_date, total_customers, SKU, price, amountLeft, containersBought)
            k += 1
        total_customers  += 1
        customer_number = customer_number + 1
    if current_date == simulation_end_date:
        break
    current_date += datetime.timedelta(1)
con.commit()
```
